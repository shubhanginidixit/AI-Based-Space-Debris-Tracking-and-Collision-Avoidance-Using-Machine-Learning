# Exploratory Data Analysis (EDA)
In this notebook, we will explore the cleaned Space-Track dataset. We'll analyze the distributions of essential orbital parameters, check for any anomalies, and understand relationships between different features through visual means.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Set plot styles
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette('viridis')

# Load data
data_path = os.path.join('..', 'data', 'processed', 'space_track_cleaned.csv')
df = pd.read_csv(data_path)
df['EPOCH'] = pd.to_datetime(df['EPOCH'])

df.head()
```

## 1. Dataset Summary & Missing Value Analysis

```python
print("Dataset Shape:", df.shape)
print("\n--- Dataset Info ---")
df.info()

print("\n--- Missing Values ---")
print(df.isnull().sum())
```

**Explanation**: Here we confirm that our dataset has 14 features and 0 missing values, as expected from our data cleaning pipeline. The data types are properly formatted (numeric and datetime).

## 2. Mean Motion Distribution

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['MEAN_MOTION'], bins=50, kde=True)
plt.title('Distribution of Mean Motion (Revs per Day)')
plt.xlabel('Mean Motion')
plt.ylabel('Frequency')
plt.show()
```

**Explanation**: Mean Motion indicates how many orbits a satellite completes per day. We can see distinct peaks. A large peak around 14-15 indicates LEO (Low Earth Orbit) satellites which orbit rapidly. A peak near 1 indicates GEO (Geosynchronous Earth Orbit) satellites which orbit once per day.

## 3. Inclination Distribution

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['INCLINATION'], bins=50, kde=True, color='coral')
plt.title('Distribution of Inclination (Degrees)')
plt.xlabel('Inclination')
plt.ylabel('Frequency')
plt.show()
```

**Explanation**: Inclination is the angle between the orbital plane and the Earth's equator. The distribution shows high frequencies at specific bands (e.g., near 98 degrees for Sun-Synchronous Orbits, or near 0 for Equatorial orbits).

## 4. Eccentricity Distribution

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['ECCENTRICITY'], bins=50, kde=True, color='purple')
plt.title('Distribution of Eccentricity')
plt.xlabel('Eccentricity')
plt.ylabel('Frequency')
plt.show()
```

**Explanation**: Eccentricity defines how circular or elliptical an orbit is (0 is perfectly circular, >0 is elliptical). The heavy skew towards 0 indicates that most tracked objects (especially functional satellites and debris in LEO) have near-circular orbits.

## 5. Scatter Plot: Mean Motion vs. Eccentricity

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='MEAN_MOTION', y='ECCENTRICITY', data=df, alpha=0.5)
plt.title('Mean Motion vs. Eccentricity')
plt.xlabel('Mean Motion (Revs per Day)')
plt.ylabel('Eccentricity')
plt.show()
```

**Explanation**: This scatter plot helps cluster orbits into regimes. For example, high mean motion (~14) with near-zero eccentricity corresponds to LEO. Lower mean motion with varying eccentricities can correspond to GTO (Geostationary Transfer Orbits) or MEO (Medium Earth Orbits).

## 6. Correlation Matrix

```python
plt.figure(figsize=(12, 8))
# Select only numerical features for correlation
num_cols = df.select_dtypes(include=[np.number]).columns.tolist()
corr = df[num_cols].corr()

sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Matrix of Orbital Parameters')
plt.show()
```

**Explanation**: The correlation heatmap reveals strong mathematical relationships. For instance, `PERIOD` and `MEAN_MOTION` have a strong negative correlation (higher mean motion means shorter orbital period). `APOAPSIS` and `SEMIMAJOR_AXIS` are strongly positively correlated. These correlations are critical to know before feeding data into our LSTM, as highly correlated features might be redundant.
